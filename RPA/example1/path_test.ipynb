```python
import os
import shutil
```

```python
PDF_DIR='./pdf_files/'
# constant variable (상수 = 변하지 않는)
```

```python
# 변수
file_list=os.listdir(PDF_DIR)
```

```python
file_list
```

```python
for i in file_list:
    # 원래 파일 위치
    original_dir=PDF_DIR+i

    # 최종 파일 위치
    target_dir=PDF_DIR+i.split('_')[1]+'/'+i

    try:
        os.makedirs(PDF_DIR+i.split('_')[1])
        # 파일 이동
        shutil.move(original_dir,target_dir)
    except:
        # 파일 이동
        shutil.move(original_dir,target_dir)
        pass

```
