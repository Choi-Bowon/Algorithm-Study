```python
import os
import shutil
```

```python
PDF_DIR='./pdf_files'

file_list=os.listdir(PDF_DIR)
```

```python
# import re
for file in file_list:
    # print(re.search('(?<=\_)[^_]*(?=\_)',file).group())
   
    # 파일이름에서 _ 기준으로 문자열 분리
    folder_name=file.split('_')[1]
    
    #  폴더를 생성해준다

    print(file.split('_')[1])
    # file.split('')
    break
```

```python
# 폴더 생성 (예외처리)
try:
    os.makedirs('./test')
except:
    pass

# os.path.isdir()

# 폴더가 이미 존재하면,
if os.path.isdir(PDF_DIR+folder_name):
    pass
# 존재하지 않으면,
else:
    os.makedirs(PDF_DIR+folder_name)

original_dir=PDF_DIR+file
target_dir=PDF_DIR+folder_name+'/'+file

# pdf_files 아래에 있는 파일들을 => original_dir
# pdf_files의 연월일 폴더에 이동한다. => target_dir
shutil.move(original_dir,target_dir)
```
