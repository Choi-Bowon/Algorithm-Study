```python
result_list=[]

for j in range(1,6):
    parameters='pageNo='+str(j)
    URL=BASE_URL+parameters

    response=requests.get(URL)
    events=response.json()['event']
    try:
        for i in range(len(events)):

            event_title=events[i]['evtNm']
            event_desc=events[i]['evtDscr']
            event_period=events[i]['evtBegDt']+'~'+events[i]['evtEndDt']
            event_link=f'https://www.eqlstore.com/special/{events[i]["evtNo"]}/view'

            event_dict={
                '제목':event_title,
                '설명':event_desc,
                '기간':event_period,
                '링크 바로가기':event_link
            }
            result_list.append(event_dict)
    except:
        pass

result_list
    # for event in events:
    #     event['evtNm']
    # for event in events:
```

```python
import time

while True:
    try:
        # 신규 메세지가 없을 경우 에러가 발생
        # list index out of range
        # 따라서, try - except 문으로 묶어줌
        new_message=bot.get_updates(offset=last_id)[-1]

        bot.send_message(chat_id,'이벤트')
        # 만약 신규 메시지가 이벤트면,
        if new_message.message.text=='이벤트':
            # 관련 메시지 발송
            bot.send_message(chat_id,result_list[0])
        # 만약 신규 메시지가 날씨이면,
        elif new_message.message.text=='날씨':
            # 관련 메시지 발송
            bot.send_message(chat_id,'오늘 날씨는 화창합니다.')
        # offset 값 최신화 (update_id) + 1 해줘서 그 다음 메시지부터 확인하도록
        last_id=new_message.update_id+1
    except:
        pass
    
    # 텔레그램 서버 부하 줄이기 위해 3초마다 확인
    time.sleep(3)
```
